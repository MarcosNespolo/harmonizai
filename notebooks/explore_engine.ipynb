Este notebook permite testar interativamente a recomendação final de vinhos a partir de uma frase em linguagem natural.

```python
# Isso força o Jupyter a recarregar automaticamente os arquivos .py do disco!
%load_ext autoreload
%autoreload 2

import sys
import os
import json

# Adiciona a raiz do projeto ao path do sistema de forma absoluta para evitar erros no Jupyter
root_dir = r'c:\Users\marco\OneDrive\Documentos\HarmonizaAi'
if root_dir not in sys.path:
    sys.path.insert(0, root_dir)

from src.nlp.pipeline import FoodMatcher
from src.engine.recommender import RecommendationEngine
```

```python
# Inicializa os componentes (pode demorar alguns segundos na primeira vez)
matcher = FoodMatcher()
engine = RecommendationEngine()
```

```python
query = "quero comer fondue de queijo hoje a noite"

# 1. NLP
nlp_result = matcher.match(query)

if nlp_result["matched_dishes"]:
    best_match = nlp_result["matched_dishes"][0]
    dish_id = best_match["id"]
    print(f"Prato identificado pelo NLP: {dish_id} (Confiança: {best_match['confidence']:.2f})")
    
    dish_data = next((d for d in matcher.dishes if d["id"] == dish_id), None)
    
    # 2. Recomendação
    recommendations = engine.recommend(dish_data, limit=5)
    
    print("\n--- TOP 5 VINHOS RECOMENDADOS ---")
    for i, rec in enumerate(recommendations, 1):
        score = rec['score']['total_score']
        print(f"\n{i}. {rec['winery']} - {rec['name']}")
        print(f"   Avaliação: {rec['rating']} estrelas | Tipo ID: {rec['type_id']} | Score Total: {score:.3f}")
        print(f"   Detalhes: [Food: {rec['score']['components']['s_food']:.2f}] "
              f"[Flavor: {rec['score']['components']['s_flavor']:.2f}] "
              f"[Struct: {rec['score']['components']['s_structure']:.2f}] "
              f"[Rating: {rec['score']['components']['s_rating']:.2f}]")
else:
    print("Nenhum prato reconhecido. Tente outra frase!")
```
